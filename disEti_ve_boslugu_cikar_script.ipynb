```python
from skimage.util.dtype import img_as_uint
import os
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt


dataset_dir = "C:\\Users\\harun\\OneDrive\\Masaustu\\StajProjesi\\Dataset"
new_datset_dir = "C:\\Users\\harun\\OneDrive\\Masaustu\\StajProjesi\\new_improved_datasets"

def disEti_ve_boslugu_cikar(img):

  # Görüntüyü griye dönüştür
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  # Siyaha yakın olan pikselleri belirle (örneğin 50'den küçük yoğunluğa sahip pikseller)
  mask = gray < 90

  # Belirlenen pikselleri orijinal görüntüde sil (siyaha dönüştür)
  img[mask] = [0, 0, 0]

  # Görüntüyü HSV'ye dönüştür
  hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

  # Diş etlerinin rengini temsil eden HSV değerlerini belirle
  lower_pink = np.array([100, 30, 70])
  upper_pink = np.array([180, 255, 255])

  # HSV görüntüsünde belirlenen renk aralığına sahip pikselleri bul
  mask = cv2.inRange(hsv, lower_pink, upper_pink)

  # Bulunan pikselleri siyahla değiştir
  img[mask > 0] = [0, 0, 0]

  # Görüntüyü BGR'dan RGB'ye çevir
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img


def create_folder(folder_path):
    try:
        # Klasörü oluştur
        os.makedirs(folder_path)
        print(f"Klasör oluşturuldu: {folder_path}")
    except FileExistsError:
        print(f"Klasör zaten mevcut: {folder_path}")

def save_image_to_folder(image, folder_path, image_name):
    # Görüntüyü RGB'den BGR'ye çevir
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    image_path = os.path.join(folder_path, image_name)
    cv2.imwrite(image_path, image)



def imporve_images(dataset_dir, new_dataset_dir, method_to_improve, new_dataset_name):

    new_dataset_dir = new_dataset_dir + "\\" + new_dataset_name
    create_folder(new_dataset_dir)
    # alt kategorileri ele alinmasini sağlıyoruz.
    all_folders = glob.glob(os.path.join(dataset_dir, '*'))
    for folder in all_folders:
        print(new_dataset_dir)
        new_dataset_folder = new_dataset_dir + "\\" + folder[-1]
        create_folder(new_dataset_folder)
        print(new_dataset_folder)
        # Veri seti dizinindeki tüm dosya adlarını alalım
        all_files = glob.glob(os.path.join(folder, '*'))
        # fotoları tek tek gezerek iyileştirme fonksiyonuna göndererek yeni oluşan verisetinin pathine kaydediliyor.
        for file in all_files:
            img = cv2.imread(file, cv2.IMREAD_COLOR)
            if img is None:
                print(f"Hata: {file} görüntüsü yüklenemedi!")
                continue

            image = method_to_improve(img)
            save_image_to_folder(image, new_dataset_folder, os.path.basename(file))

imporve_images(dataset_dir, new_datset_dir, disEti_ve_boslugu_cikar, "disEti_ve_boslugu_cikar")
```
